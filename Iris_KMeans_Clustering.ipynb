```python
# Import dependencies
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Loading dataset
iris_df = datasets.load_iris()

# Check available methods
print(dir(iris_df))
```

```python
# Target
print(iris_df.target_names)
```

```python
# Create dataframe
df = pd.DataFrame(iris_df.data, columns = iris_df.feature_names)
```

```python
# Check all features scatterplot
sns.set_style('whitegrid')
sns.pairplot(df)
plt.show()
```

```python
# Import KMeans clustering algorithm
from sklearn.cluster import KMeans
```

```python
# Create function to show an inertias
def optimize_k_means(data, max_k):
    means = []
    inertias = []

    for k in range(1, max_k):
        model = KMeans(n_clusters = k)
        model.fit(data)
        means.append(k)
        inertias.append(model.inertia_)
    
    fig = plt.subplots(figsize = (10, 5))
    plt.plot(means, inertias, 'o-')
    plt.xlabel('K Number')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show
```

```python
# Call the function
optimize_k_means(df, 30)
```

```python
# Create model with optimum n_cluster
model = KMeans(n_clusters = 3)
model.fit(df)
```

```python
# Get the cluster labels
df['KMeans'] = model.labels_
df.head()
```

```python
sns.set_style('whitegrid')
sns.pairplot(df, hue = 'KMeans', height = 3)
plt.show()
```

```python
df.head()
```

```python
centroid = pd.DataFrame(model.cluster_centers_, columns = iris_df.feature_names)
centroid['KMeans'] = [0, 1, 2]
centroid['labels'] = ['centroid-0', 'centroid-1', 'centroid-2']
centroid
```

```python
df_all = pd.merge(df, centroid, on = 'KMeans')
df_all
```

```python
columns = list(df.columns)
columns.remove('KMeans')

fig, ax = plt.subplots(4, 4, figsize = (30, 30))

sns.set_style('whitegrid')

num_x = 0
for col_x in columns:

    num_y = 0
    for col_y in columns:
        sns.scatterplot(data = df, x = col_x, y = col_y, ax = ax[num_x, num_y], hue = 'KMeans')
        sns.scatterplot(data = centroid, x = col_x, y = col_y, ax = ax[num_x, num_y], hue = 'labels', s = 70, palette = 'tab10')

        for i in range(len(df_all[col_x + '_x'])):
            x = [df_all.loc[i, col_x + '_x'], df_all.loc[i, col_x + '_y']]
            y = [df_all.loc[i, col_y + '_x'], df_all.loc[i, col_y + '_y']]
            sns.lineplot(x, y, ax = ax[num_x, num_y])
        num_y += 1

    num_x += 1
```
